```python
import pandas as pd
```

```python
df= pd.read_csv('feo.csv', encoding='latin-1')
```

```python
df
```

```python
#convertir hielos a booleanos

df['hielo '] =(df['hielo '] == "con").astype(int)
df['tamaño'] = (df['tamaño'] == 'grande').astype(int)
df['kahlua']= (df['sabor'] == 'kahlua').astype(int)
df['horchata']= (df['sabor'] == 'horchata').astype(int)
df.drop('sabor', axis=1, inplace=True)
df
```

```python
import numpy as np
```

```python
tamaño= df['tamaño'].to_numpy().reshape(-1,1)
hielo =df['hielo '].to_numpy().reshape(-1,1)
horchata=df['horchata'].to_numpy().reshape(-1,1)
kahlua= df['kahlua'].to_numpy().reshape(-1,1)
X= np.hstack((tamaño, hielo, kahlua, horchata))
y= df['score'].to_numpy().reshape(-1,1)
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
model=sm.OLS(y, X) # -> Modela minimos cuadrados ordinarios con todas las variables
results=model.fit() #-> Optimiza
print(results.summary())
```

```python
df['hielo '] = df['hielo '].replace(0, -1)
df['tamaño'] = df['tamaño'].replace(0,-1)
df
```

```python
X = sm.add_constant(X)
model=sm.OLS(y, X) # -> Modela minimos cuadrados ordinarios con todas las variables
results=model.fit() #-> Optimiza
print(results.summary())
```
