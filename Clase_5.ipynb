```python
import pandas as pd
```

```python
df= pd.read_csv('Salarios_minimos.csv')
```

```python
df.describe()
```

```python
X=df['Periodo'].to_numpy().reshape(-1,1)
y= df['Salario'].to_numpy().reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
lr.fit(X,y)
B1= lr.coef_
B0= lr.intercept_
print(f'{B0}, {B1}')
R2= lr.score(X,y)
print(f'R2: {R2}')
```

```python
#Con periodo/1000
X_1= X/1000
```

```python
lr= LinearRegression()
lr.fit(X_1,y)
B1= lr.coef_
B0= lr.intercept_
print(f'{B0}, {B1}')
R2= lr.score(X_1,y)
print(f'R2: {R2}')
```

```python
raw_df2 = pd.read_csv('Boston Housing Dataset 1978.csv')
raw_df2
```

```python
def normalizar():
  for i in range(len(raw_df2.columns)):
    raw_df2.iloc[:,i]= (raw_df2.iloc[:,i]-raw_df2.iloc[:,i].min())
    raw_df2.iloc[:,i]= raw_df2.iloc[:,i]/raw_df2.iloc[:,i].max()
  return raw_df2
```

```python
normalizar()
```

```python
def normalizar_z(raw_df2):
  for i in range(len(raw_df2.columns)):
    raw_df2.iloc[:,i]= (raw_df2.iloc[:,i]-raw_df2.iloc[:,i].mean())
    raw_df2.iloc[:,i]= raw_df2.iloc[:,i]/raw_df2.iloc[:,i].std()
  return raw_df2
```

```python
plt.hist(normalizar_z(raw_df2))
```
